Задание 2. График 2.7.

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.figure(figsize=(10, 5))
x_positions = [0, 0.75, 3, 5, 7, 9]
for i in range(len(x)):
  plt.bar(x_positions[i], y[i], color='darkblue', width=0.5)

plt.xticks([pos + 0.25 for pos in x_positions], x)
plt.grid(color='red')
plt.grid(which="minor")
plt.minorticks_on()
plt.title('Popularity of Programming Language\nWorldwide, Oct 2017 compared to a year ago')
plt.xlabel('Languages')
plt.ylabel('Popularity')
```

Задание 1. Построить несколько графиков на одном поле

Исследуем распределение витаминов

```python
fig, axes = plt.subplots(6, 1, figsize=(10, 10), constrained_layout=True)

sns.boxplot(data=df, y='Data.Vitamins.Vitamin A - RAE', color='red',width=0.2, ax=axes[0], showfliers=False)
axes[0].set_title('Гистограмма распределения витамина А')
axes[0].set_ylabel('Витамин А')

sns.boxplot(data=df, y='Data.Vitamins.Vitamin B12', color='green',width=0.2, ax=axes[1], showfliers=False)
axes[1].set_title('Гистограмма распределения витамина B12')
axes[1].set_ylabel('Витамин B12')

sns.boxplot(data=df, y='Data.Vitamins.Vitamin B6', color='orange',width=0.2, ax=axes[2], showfliers=False)
axes[2].set_title('Гистограмма распределения витамина B6')
axes[2].set_ylabel('Витамин B6')

sns.boxplot(data=df, y='Data.Vitamins.Vitamin C', color='yellow',width=0.2, ax=axes[3], showfliers=False)
axes[3].set_title('Гистограмма распределения витамина С')
axes[3].set_ylabel('Витамин C')

sns.boxplot(data=df, y='Data.Vitamins.Vitamin E', color='purple',width=0.2, ax=axes[4], showfliers=False)
axes[4].set_title('Распределение витамина E')
axes[4].set_ylabel('Витамин E')

sns.boxplot(data=df, y='Data.Vitamins.Vitamin K', color='blue',width=0.2, ax=axes[5], showfliers=False)
axes[5].set_title('Распределение витамина К')
axes[5].set_ylabel('Витамин К')

plt.show()
```

Задание 2. Построить диаграммы для категориальных данных

В датасете есть только один категоиальный признак - категория

Создадим ограниченный датафрейм на 50 первых записей, так как категорий слишком много из-за огромного размера первоначального датасета.

```python
dflim=pd.read_csv("food.csv", nrows=50)
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x='Category', data = dflim)
plt.title('Диаграмма для категориального поля')
plt.xlabel('Категория')
plt.ylabel('Количество записей')
```

Задание 3. Построить столбчатые и круговые диаграммы

Столбчатая диаграмма:

```python
fig=plt.subplots(figsize=(12, 6))
sns.countplot(data=df, x='Data.Major Minerals.Calcium')
plt.title('Столбчатая диаграмма по распределению кальция')
plt.xlabel('Количество кальция')
plt.ylabel('Количество продуктов')
plt.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False,
    labelbottom=False)
plt.show()
```

```python
largest5 = df['Category'].value_counts().nlargest(5)

plt.figure(figsize=(8, 8))
plt.pie(largest5, labels=largest5.index, autopct='%1.1f%%')
plt.title('Круговая диаграмма 5-ти самых больших категорий')
plt.show()
```

Задание 4. Реализовать цветовую сетку данных

```python
cormat = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(cormat, square=True)
plt.title('Цветовая сетка')
```

Задание 5. Построить 3D-график

```python
fig = plt.figure(figsize=(10, 7))
axes = fig.add_subplot(111, projection='3d')
x = df['Data.Thiamin']
y = df['Data.Sugar Total']
z = df['Data.Water']

axes.scatter(x, y, z)
plt.title('3D-график данных распределения продуктов по трем веществам')
axes.set_xlabel('Тиамин')
axes.set_ylabel('Сахар')
axes.set_zlabel(' Вода ')

plt.show()
```
