12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
FMat=np.random.rand(5,5)
Mmax=np.max(FMat, axis=1, keepdims=True)
Mmin=np.min(FMat, axis=1, keepdims=True)
NFMat=np.hstack((FMat, Mmax))
NNFMat=np.hstack((NFMat, Mmin))
print('Исходная матрица: \n ', FMat)
print('\nМатрица после добавления столбцов: \n', NNFMat)
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.  

```python
IVec=np.arange(1,52)

EVec=IVec[IVec%2 ==0]
print('\nВектор четных чисел: \n', EVec)
print('\nКоличество элементов: {}'.format(EVec.size))

OVec=IVec[IVec%2 !=0]
OOVec=OVec[0:25]
print('\nВектор нечетных чисел (с измененной размерностью): \n', OOVec)
print('\nКоличество элементов: {}'.format(OOVec.size))

EMat=EVec.reshape(5,5)
print('\nМатрица четных чисел: \n', EMat)

OMat=OOVec.reshape(5,5)
print('\nМатрица нечетных чисел: \n', OMat)

PMat=np.dot(EMat,OMat)
print('\nПроизведение матриц: \n', PMat)

print('\nОбратная матрица матрицы произведения: ')
try:
  InvMat= np.linalg.inv(PMat)
  print('\nобратная матрица: \n', InvMat)
except:
  print('обратной матрицы не сущесвует')
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
ZMat=np.zeros((8,8), int)
print('\nИсходная матрица: \n', ZMat)
ZMat[::2,::2]=1
print('\nПервое заполнение единицами: \n', ZMat)
ZMat[1::2,1::2]=1
print('\nВторое заполнение единицами: \n', ZMat)
```
